Вместо использования `matplotlib` можно также строить графики с помощью `plotly`.

```python
# Вначале:

import scipy.stats as sps
sns.set(font_scale=1.6, palette='autumn')
k = 500
color = ['red','orange']
for n in (10, 100, 500):
    plt.figure(figsize=(12, 3))
    arr = np.array(sps.norm.rvs(size=n * k))
    arr.shape = (n, k)
    arr2 = arr * arr
    arr = arr.mean(axis=0)
    arr2 = arr2.mean(axis=0)
    theta = -arr * arr +  arr2
    # Для каждой оценки:
    plt.scatter(theta, np.zeros(k) + 0,
                alpha=0.1, s=100, color=color[0], label='оценка сигмы - S^2')
    plt.scatter(theta.mean(), 0, marker='*', s=200, 
                color='w', edgecolors='black')
    arr = np.array(sps.norm.rvs(size=n * k))
    arr.shape = (n, k)
    arr2 = arr * arr
    arr = arr.mean(axis=0)
    arr2 = arr2.mean(axis=0)
    theta = n/(n-1) * ((-1) * arr * arr +  arr2)
    # Для каждой оценки:
    plt.scatter(theta, np.zeros(k) + 1,
                alpha=0.1, s=100, color=color[1], label='оценка сигмы - (n/n-1) * S^2')
    plt.scatter(theta.mean(), 1, marker='*', s=200, 
                color='w', edgecolors='black')
    # Для всего графика:
    plt.vlines(1, 0, 1, color='r')
    plt.title(f'Размер выборки = {k}')
    plt.yticks([])
    plt.legend(loc='right')
    plt.show()
```

**2.** Изучим поведение среднего оценок из первого пункта при росте размера выборки. Постройте график зависимости $\widetilde{\theta}$ от $n$ для двух типов оценок. Для вычисления зависимости нужно один раз сгенерировать выборки из пункта 1 достаточно большого размера (не более 500) и посчитать оценки по подвыборкам, используя <a href="https://mipt-stats.gitlab.io/courses/python/05_numpy.html#2.1-Математические-операции">функции</a> из `numpy`. Использовать циклы, а так же функции, разворачивающиеся в цикл (например, `np.vectorize`), запрещено. 

**Решение:**

```python
plt.figure(figsize=(12, 7))
plt.ylim((0.95, 1.05))
n = 500
k = 500
arr = np.array(sps.norm.rvs(size=n * k))
arr.shape = (k, n)
arr2 = arr * arr
arr = arr.cumsum(axis=1)
arr2 = arr2.cumsum(axis=1)
arr = arr / np.arange(1, n + 1)
arr2 = arr2 / np.arange(1, n + 1)
theta = arr2 - arr * arr
theta = theta.mean(axis = 0)
sns.lineplot(y=theta, x=np.arange(1, n + 1), label='сигма оценивается как S^2')
arr = np.array(sps.norm.rvs(size=n * k))
arr.shape = (k, n)
arr2 = arr * arr
arr = arr.cumsum(axis=1)
arr2 = arr2.cumsum(axis=1)
arr = arr / np.arange(1, n + 1)
arr2 = arr2 / np.arange(1, n + 1)
theta = arr2 - arr * arr
theta = theta.mean(axis = 0)
koef = np.arange(2, n + 1) / (np.arange(2, n + 1) - 1)
sns.lineplot(y=theta[1:] * koef, x=np.arange(2, n + 1), color='green', label='сигма оценивается как (n/n-1) * S^2')
plt.title('Среднее значение оценки, посчитанной на n значениях')
plt.ylabel('Среднее значение оценки')
plt.xlabel('Значение n')
plt.show()
```

Поясните, почему в лабораторных по физике при оценке погрешности иногда используют $n-1$ в знаменателе, а не $n$.

**Вывод:** 

Несмещенность оценки - свойство, которое означает, что какое бы мы не взяли тетта, мат ожидание с параметром тетта от нашей функции (в данном случае, эта функция - оценка параметра) будет равна тетте. Несмещенность оценки гарантирует нам, что при фиксированном n у нас не будет появлятся "ошибки в среднем". При n-> inf, и у смещенной, и у несмещенной оценки погрешности будут минимальны, однако, даже при больших n у смещенной оценки будет наблюдаться отклонение, какая-то функция от n. Поэтому на практике лучше использовать несмещенную оценку, так как при маленьких n (как например, в лабораторных по физике), это смещение будет ощутимым. Поэтому там используют n - 1 в знаменателе, так как эта оценка параметров нормального распределения несмещена. На графике видно, что несмещенная оценка гораздо лучше себя ведет при малых n (< 100)
